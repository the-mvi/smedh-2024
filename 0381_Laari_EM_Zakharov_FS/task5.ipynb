##–¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–û–∑–Ω–∞–∫–æ–º–ª–µ–Ω–∏–µ —Å –æ—Å–Ω–æ–≤–Ω—ã–º–∏ –ø–æ–ª–æ–∂–µ–Ω–∏—è–º–∏ –º–µ—Ç–æ–¥–∞ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ (–ú–ù–ö), —Å–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–º–∏ —Å–≤–æ–π—Å—Ç–≤–∞–º–∏ –ú–ù–ö –æ—Ü–µ–Ω–æ–∫, —Å –ø–æ–Ω—è—Ç–∏–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ —Ä–æ–ª–∏ –ú–ù–ö –≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–æ–Ω–Ω–æ–º –∞–Ω–∞–ª–∏–∑–µ, —Å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—ã–º –æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º, –∫–∞–∫ –º–µ—Ä–æ–π —Ç–µ—Å–Ω–æ—Ç—ã –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π (–≤ —Ç–æ–º —á–∏—Å–ª–µ –∏ –ª–∏–Ω–µ–π–Ω–æ–π) –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π —Å–≤—è–∑–∏.

##–û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è.
–ú–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ (–ú–ù–ö) - –º–µ—Ç–æ–¥, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã–π –Ω–∞ –ø–æ–∏—Å–∫–µ
–º–∏–Ω–∏–º—É–º–∞ —Å—É–º–º—ã –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –∑–Ω–∞—á–µ–Ω–∏–π –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –æ—Ç
–∑–∞–¥–∞–Ω–Ω–æ–≥–æ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π.
–§—É–Ω–∫—Ü–∏—è $ùëî(ùë•)$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π —Å—Ä–µ–¥–Ω–µ–π
–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ Y –Ω–∞ X
$$g(x) = m(Y)+r_{xy}\frac{\sigma_y}{\sigma_x}(x-m(X))$$

–î–ª—è —Ä–∞—Å—á–µ—Ç–æ–≤ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ñ–æ—Ä–º—É–ª–∞ —Å–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–º–∏ –æ—Ü–µ–Ω–∫–∞–º–∏
–Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤:
$$\overline{y}_x = \overline{y}_B+\overline{r}_{xy}\frac{S_y}{S_x}(x - \overline{x}_B)$$

–ê–Ω–∞–ª–æ–≥–∏—á–Ω—ã–º –æ–±—Ä–∞–∑–æ–º —Å—Ç—Ä–æ—è—Ç—Å—è —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è $\overline{x}_y\$

–í–Ω—É—Ç—Ä–∏–≥—Ä—É–ø–ø–æ–≤–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è - –≤–∑–≤–µ—à–µ–Ω–Ω–æ–µ –ø–æ –æ–±—ä–µ–º–∞–º –≥—Ä—É–ø–ø —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≥—Ä—É–ø–ø–æ–≤—ã—Ö –¥–∏—Å–ø–µ—Ä—Å–∏–π.

–ú–µ–∂–≥—Ä—É–ø–ø–æ–≤–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è - –¥–∏—Å–ø–µ—Ä—Å–∏—è —É—Å–ª–æ–≤–Ω—ã—Ö (–≥—Ä—É–ø–ø–æ–≤—ã—Ö)
—Å—Ä–µ–¥–Ω–∏—Ö ${\overline{x}_y}_i$ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ $\overline{x}_B$.

–û—Ü–µ–Ω–∫–∞ –æ–±—â–µ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ X –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç—Å—è –∫–∞–∫:
$$D_{–æ–±—â} = D_{–º–µ–∂–≥—Ä} +D_{–≤–Ω–≥—Ä}$$

–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ X –∫ Y —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç $\overline{∆û}_{xy}$:
$$\overline{∆û}_{xy}= \frac{\overline{\sigma}_{\overline{x}_y}}{\overline{\sigma}_x}$$
$$\overline{\sigma}_{\overline{x}_y} = \sqrt{D_{–º–µ–∂–≥—Ä}}$$
$$\overline{\sigma}_x = \sqrt{D_{–æ–±—â}}$$

–ê–Ω–∞–ª–æ–≥–∏—á–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –¥–ª—è $\overline{Œ∑}_{yx}$ . –¢–∞–∫–∂–µ –∏–º–µ–µ—Ç –º–µ—Å—Ç–æ —Å–ª–µ–¥—É—é—â–µ–µ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ:
$$D_{–≤–Ω–≥—Ä} = D_{–æ–±—â}(1-Œ∑^2)$$

–°–≤–æ–π—Å—Ç–≤–∞ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—è:
1. $0 ‚â§ \overline{Œ∑} ‚â§ 1 $
2. $\overline{Œ∑} = 0$ - –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–æ–≤–∞–Ω—ã —Å –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ, —á—Ç–æ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã X –∏ Y –Ω–µ —Å–≤—è–∑–∞–Ω—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å;
3. $\overline{Œ∑} = 1$ - –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–æ–≤–∞–Ω—ã —Å –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ–º, —á—Ç–æ
—Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã X –∏ Y —Å–≤—è–∑–∞–Ω—ã —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å—é;
4. $\overline{Œ∑} ‚â• |\overline{r}_{xy}|$
5. $\overline{Œ∑} = |\overline{r}_{xy}|$ - –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–æ–≤–∞–Ω—ã —Å –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ–º,
—á—Ç–æ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã X –∏ Y —Å–≤—è–∑–∞–Ω—ã –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π
–∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å—é;
6. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —è–≤–ª—è–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –º–µ—Ä–æ–π
—Ç–µ—Å–Ω–æ—Ç—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –º–µ–∂–¥—É —Å–ª—É—á–∞–π–Ω—ã–º–∏ –≤–µ–ª–∏—á–∏–Ω–∞–º–∏ X –∏
Y. –ù–æ –æ–Ω–æ –Ω–µ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏
(–∫—Ä–æ–º–µ –ø—É–Ω–∫—Ç–∞ 5).

2. –î–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å —É—Ä–∞–≤–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–π –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ x –Ω–∞ y –∏ y –Ω–∞ x —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –ø—Ä—è–º—ã–µ –Ω–∞ –º–Ω–æ–∂–µ—Å—Ç–≤–µ –≤—ã–±–æ—Ä–∫–∏. –û–±—ä—è—Å–Ω–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

$\overline y_x=\overline y + \overline r \frac{S_y}{S_x}(x - \overline x)$

$\overline x_y=\overline x + \overline r \frac{S_x}{S_y}(y - \overline y)$

$\overline y=127.34$

$\overline x=449.79$

$S_x=57.11$

$S_y=23.29$

$\overline r=0.9$

$\overline y_x=127.34 + 0.9 \frac{23.29}{57.11}(x - 449.79)$

$\overline x_y=449.79 + 0.9 \frac{57.11}{23.29}(y - 127.34)$

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

overline_x = sum(data['A']) / N
overline_y = sum(data['B']) / N
S_x = 57.11
S_y = 23.29
r = 0.9

min_val, max_val = min(data['B']), max(data['B'])
x_minval = overline_x + r * S_x / S_y * (min_val - overline_y)
x_maxval = overline_x + r * S_x / S_y * (max_val - overline_y)
ax.plot([x_minval, x_maxval], [min_val, max_val], "-b", label="$x_y$ = 449.79 + 2.207(y-127.34)")

min_val, max_val = min(data['A']), max(data['A'])
x_minval = overline_y + r * S_y / S_x *(min_val - overline_x)
x_maxval = overline_y + r * S_y / S_x *(max_val - overline_x)
ax.plot([min_val, max_val], [x_minval, x_maxval], "-r", label="$y_x$ = 127.34 + 0.367(x-449.79)")

ax.legend()
ax.set_title('–ü—Ä—è–º—ã–µ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏')
plt.scatter(data["A"], data["B"], c ="black")
```

–ë—ã–ª–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω—ã –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏ $y(x)$ –∏ $x(y)$, —è–≤–ª—è—é—â–∏–µ—Å—è –±–∞–∑–æ–≤—ã–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π —Å–≤—è–∑–∏, –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—â–∏–º –µ–µ –æ—Å–Ω–æ–≤–Ω—É—é —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é. –ö–∞–∫ –≤–∏–¥–Ω–æ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ, –æ–±–µ –ø—Ä—è–º—ã–µ —É–≤–µ–ª–∏—á–∏–≤–∞—é—Ç—Å—è, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—É—é –¥–∏–Ω–∞–º–∏–∫—É: —á–µ–º –≤—ã—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ X, —Ç–µ–º –≤—ã—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ Y.


3. –°–æ—Å—Ç–∞–≤–∏—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—è. –£–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ $Œ∑_{xy}‚©æ|r_{xy}|$ –∏ $Œ∑_{yx}‚©æ|r_{xy}|$ –≤—ã–ø–æ–ª–Ω—è—é—Ç—Å—è.

```python
import math
intervals_numberA = int(1 + 3.322 * math.log10(N))
min_densityA, max_densityA = min(data["A"]), max(data["A"])
range_densityA = max_densityA - min_densityA

intervals_numberB = int(1 + 3.322 * math.log10(N))
min_densityB, max_densityB = min(data["B"]), max(data["B"])
range_densityB = max_densityB - min_densityB
```

```python
isInIntervalA = lambda x: min(int((abs(x) - min_densityA) / range_densityA * intervals_numberA), intervals_numberA - 1)
bordersA = [((min_densityA + range_densityA) / intervals_numberA * i, (min_densityA + range_densityA) / intervals_numberA * (i + 1))
           for i in range(intervals_numberA)]
intervalsA = [[] for i in range(intervals_numberA)]

for value in data["A"]:
    intervalsA[isInIntervalA(value)].append(value)

intervalsA = [sorted(i) for i in intervalsA]
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ A:", len(intervalsA))
print(intervalsA)

isInIntervalB = lambda x: min(int((abs(x) - min_densityB) / range_densityB * intervals_numberB), intervals_numberB - 1)
bordersB = [((min_densityB + range_densityB) / intervals_numberB * i, (min_densityB + range_densityB) / intervals_numberB * (i + 1))
           for i in range(intervals_numberB)]
intervalsB = [[] for i in range(intervals_numberB)]

for value in data["B"]:
    intervalsB[isInIntervalB(value)].append(value)

intervalsB = [sorted(i) for i in intervalsB]
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤ B:", len(intervalsB))
print(intervalsB)
```

$$D_{–º–µ–∂–≥—Ä} = \frac{\sum^m_{i=1}[{n_y}_i({\overline{x}_y}_i - {\overline{x}_B})^2]}{N}$$

–í–Ω—É—Ç—Ä–∏–≥—Ä—É–ø–ø–æ–≤–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è:
$$D_{–≤–Ω–≥—Ä} = \frac{\sum^m_{i=1}{n_y}_i{D_y}_i}{N}$$

–¢–æ–≥–¥–∞ –æ–±—â–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è:
