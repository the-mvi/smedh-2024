## Основные теоретические положения
> *__Выборка__* - совокупность случайно отобранных элементов из генеральной совокупности.

> *__Репрезентативность (представительность) выборки__* <br>
&emsp;Главный принцип формирования репрезентативной выборки - *случайность ее формирования*.

> *__Генеральная совокупность__* - совокупность всех объектов, из которых производится выборка.

<br>

&emsp;*__Классификация выборок__*
* &emsp;*Повторная* - если перед выбором следующего объекта случайно отобранные объекты возвращаются в генеральную совокупность.
* &emsp;*Бесповторная* - если случайно отобранные объекты не возвращаются в генеральную совокупность.

&emsp;*__Способы формирования выборки__*:

&emsp;&emsp;*__Без разбиения генеральной совокупности на части__*:

* &emsp;*Простой случайный бесповторный отбор* - случайным образом отбираются элементы генеральной совокупности и *не возвращаются* в генеральную совокупность.

* &emsp;*Простой случайный повторный отбор*  - случайным образом отбираются элементы генеральной совокупности и *возвращаются* в генеральную совокупность.

&emsp;&emsp;*__С разбиением генеральной совокупности на части:__*

* &emsp;*Типический* - генеральная совокупность разделяется на несколько подгрупп в соответствии с каким-либо признаком, а затем производится случайный отбор в каждой подгруппе пропорциональной доли данной подгруппы от всей генеральной совокупности.

* &emsp;*Механический* - генеральная совокупность разделятся на несколько равных (по возможности) интервалов, затем из первого интервала случайно отбирается один элемент, а потом с шагом, равным длине интервала, отбираются другие элементы из последующих интервалов.

* &emsp;*Серийный* - генеральная совокупность разделяется на несколько групп (серий) таким образом, чтобы в каждой группе было поровну (по возможности) элементов; затем случайным образом выбирается одна серия, а затем при помощи простого, типического или механического отбора отбирается определенное числое элементов.

<br>

> *__Статистический ряд__* - последовательность элементов выборки, расположенных в порядке их получения (наблюдения).


> *__Ранжированный ряд__* - последовательность элементов выборки, расположенных в порядке возрастания значения исследуемого признака.
* Номер элемента ранжированного ряда в последовательности называется **рангом**.
<br><br>

> *__Вариационный ряд__* - получается из ранжированного ряда в результате объединения одинаковых элементов.
* Элементы вариационного ряда называются *__вариантами__*.
<br><br>

&emsp;*__Построение интервального ряда:__*

&emsp;Если количество вариант $K$ достаточно велико, то необходимо разделить вариационный ряд на интервалы. В этом случае ряд называется *интервальным*. Количество интервалов $k$ зависит от объема выборки $N$ и может быть вычислено по формуле Стерджесса:
$$k = 1+3.31*lgN$$
Полученное значение округляется до целого. Рекомендуется выбирать нечетное количество интервалов. Затем вычисляется ширина интервалов $h$ по формуле:
$$h = \frac{x_{max}-x_{min}}{k}$$

&emsp;*__Границы интервалов__* ($i$ - номер интервала):
$$[x_{min}+(i-1)*h; x_{min}+i*h), i = 1,2,...,k-1,$$

&emsp;*__Последний интервал__*:
$$[x_{min} + (k-1)*h; x_{min}+k*h]$$

Затем подсчитывают частоту попадания значений вариант в интервалы и вычисляют: <br>&emsp;&emsp;$\widetilde{x_i}$ - *__середины интервалов__* для $i=\overline{1, k}$
<br><br>

&emsp;*__Построение полигона и гистограммы абсолютных/относительных частот__*:

&emsp;На основании полученного интервального ряда могут
быть построены:
- *полигон абсолютных/относительных частот*
- *гистограмма абсолютных/относительных частот.*

> *__Абсолютная частота в интервальном ряде__* - количество вариант, попавших в определенный интервал интервального ряда.

> *__Относительная частота в интервальном ряде__* - отношение абсолютной частоты к общему количеству вариант.

> *__Полигон абсолютных/относительных частот__* - ломаная, соединяющая точки, соответствующие по оси Ox - серединам интервалов интервального ряда, по оси Oy - абсолютным/относительным частотам этих интервалов.

> *__Гистограмма абсолютных/относительных частот__* - фигура, состоящая из прямоугольников, основаниями которых служат интервалы интервального ряда, а высоты равны отношению абсолютных/относительных частот интервала к его ширине.
Площадь таких прямоугольников равна абсолютной/относительной частоте.
<br><br>

&emsp;*__Эмпирическая функция распределения__*:

> *__Эмпирической функцией распределения__* (функцией распределения выборки) называют функцию $F^*(x)$, определяющую для каждого значения $x$ относительную частоту события $X < x$.

&emsp;График $F^*(x)$ представляет собой лестничный график, длина каждой ступеньки которого равна длине соответствующего интервала, а высота - отношению накопленной частоты до середины этого интервала к объему выборки, т.е.
$$F^*( \widetilde{x}_i) = \frac{m_i^{нак}}{N}; \\  m_i^{нак} = \sum_{j=1}^{i-1}m_j; \\ i=1,2,...,k+1$$

### *__Задание 4__*
*Последовательно преобразовать выборку в ранжированный, вариационный и интервальный ряды. Интервальный ряд представить в виде табл. 1. Результаты содержательно проинтерпретировать.*


*__Ранжированные ряды:__*

&emsp;Было получено преобразование выборки в *__ранжированные ряды__* по исследуемым признакам:
* `pandas.DataFrame.sort_values(by='<название столбца>')` - сортирует выборку по значениям в указанном столбце в порядке возрастания.

*__Вывод по ранжированным рядам:__* <br>
&emsp;Благодаря ранжированным рядам можно выявить *__наибольшее и наименьшее значение $x_{max} \ и \ x_{min} $__* в выборке для каждого столбца, а также ранг, то есть порядковый номер, конкретного элемента в этом ряду.

```python
# Ранжированный ряд по признаку nu
ranked_df_nu = sample_df.sort_values(by='nu')
# Добавление колонки с рангом
ranked_df_nu['rank']=list(range(0, sample_df.shape[0]))
# Ранжированный ряд по признаку E
ranked_df_e = sample_df.sort_values(by='E')
# Добавление колонки с рангом
ranked_df_e['rank']=list(range(0, sample_df.shape[0]))


print('Ранжированный ряд - по признаку nu:\n {}'.format(ranked_df_nu))
# Вычисление наибольшего и наименьшего значения в выборке
x_max_nu = max(ranked_df_nu['nu'])
x_min_nu = min(ranked_df_nu['nu'])
print('Наибольшее и наименьшее значение: x_max={}, x_min={}\n'.format(x_max_nu, x_min_nu))


print('Ранжированный ряд - по признаку E:\n {}'.format(ranked_df_e))
# Вычисление наибольшего и наименьшего значения в выборке
x_max_e = max(ranked_df_e['E'])
x_min_e = min(ranked_df_e['E'])
print('Наибольшее и наименьшее значение: x_max={}, x_min={}\n'.format(x_max_e, x_min_e))
```

*__Вариационные ряды:__*

&emsp;Было получено преобразование *__ранжированного ряда__* в вариационный ряд путем удаления дублирующихся элементов в выборке:
* `pandas.DataFrame.drop_duplicates(keep='first')` - возвращает фрейм данных с удаленными повторяющимися строками.

*__Вывод по вариационным рядам:__* <br>
&emsp;Получили, что количество вариант в вариационных рядах немного меньше объема выборки (при объеме выборки 117: 87 вариант для nu и 107 вариант для E). Что говорит о том, что в выборке присутствовали дублирующиеся значения, но в малом количестве относительно объема выборки.

```python
variation_df_nu = ranked_df_nu.drop_duplicates(subset='nu', keep='first')
variation_df_e = ranked_df_e.drop_duplicates(subset='E', keep='first')

# Вариационный ряд из ранжированного ряда по признаку nu
print('Вариационный ряд из ранжированного ряда по nu:\n {}\n'.format(variation_df_nu))
#  Вариационный ряд из ранжированного ряда по признаку nu
print('Вариационный ряд из ранжированного ряда по E:\n {}'.format(variation_df_e))
```

*__Интервальные ряды:__*

&emsp;Количество вариант оказалось меньше объема выборки (т.е. дублирующиеся значения присутствовали), но количество вариант по-прежнему велико. Разделим один из получившихся вариационных рядов (для nu) на интервалы и получим *__интервальный ряд__*.

&emsp;*__Количество интервалов $k$__* вычислено по формуле Стерджесса ($N$ - объем выборки):
$$k = 1+3.31*lgN$$
Полученное значение округлили до целого так, чтобы получилось нечетное количество интервалов.

&emsp;Затем вычислили *__ширину интервалов $h$__* по формуле:
$$h = \frac{x_{max}-x_{min}}{k}$$

&emsp;*__Границы интервалов__* ($i$ - номер интервала):
$$[x_{min}+(i-1)*h; x_{min}+i*h), i = 1,2,...,k-1,$$

&emsp;*__Последний интервал__*:
$$[x_{min} + (k-1)*h; x_{min}+k*h]$$

Затем посчитали частоту попадания значений вариант в интервалы и вычислили: <br>&emsp;&emsp;$\widetilde{x_i}$ - *__середины интервалов__* для $i=\overline{1, k}$

```python
# Получение интервального ряда и Таблицы 1

# Объем выборки
N=117
# Количество интервалов по формуле Стерджесса
k = int(np.floor(1 + 3.31 * np.log10(N)))
print('Количество интервалов k={}'.format(k))

# Ширина интервалов h
h = int(np.ceil((x_max_nu - x_min_nu) / k))
print('Ширина интервалов h={}'.format(h))

# Границы интервалов
# Вставим левую границу для i=1 первого интервала
left = x_min_nu + (1-1)*h
left_borders = [left]
right_borders = []
# Вставим в списки итеративно следующие границы
for i in range(1, k):
  left_borders.append(x_min_nu + h * i)
  right_borders.append(x_min_nu + h * i)
# Добавим последнюю правую границу
right = x_min_nu + k*h
right_borders.append(right)
# Разобьем на интервалы
intervals= []
for i in range(len(left_borders)):
  intervals.append((left_borders[i], right_borders[i]))
print('Границы интервалов: {}'.format(intervals))

# Вычисляем середины интервалов
mid_of_intervals = []
for i in range(len(intervals)):
  left, right = intervals[i]
  mid_of_intervals.append((left + right) / 2)
print('Середины интервалов: {}'.format(mid_of_intervals))

# Абсолютные частоты для интервалов
abs_freq = [0 for i in range(0, k)]
# Считаем частоту попаданий вариант в каждый интервал
for varianta in variation_df_nu['nu']:
  for i in range(len(intervals)):
    left, right = intervals[i]
    if varianta < right and varianta >= left:
      abs_freq[i]+=1
print('Абсолютные частоты для интервалов: {}'.format(abs_freq))
# Количество вариант
variant_count = variation_df_nu['nu'].shape[0]
print('Кол-во вариант: {}'.format(variant_count))
# Относительные частоты для интервалов
rel_freq = [freq/ variant_count for freq in abs_freq]
print('Относительные частоты для интервалов: {}'.format(rel_freq))

# Накопленные частоты для интервалов
accum_abs_freq = []
accum_rel_freq = []
for i in range(len(abs_freq)):
  accum_abs_freq.append(sum(abs_freq[:i+1]))
  accum_rel_freq.append(sum(rel_freq[:i+1]))
print('Накопленные абсолютные частоты для интервалов: {}'.format(abs_freq))
print('Накопленные относительные частоты для интервалов: {}'.format(rel_freq))

# Добавляем необходимые колонки для Таблицы 1
interval_df = pandas.DataFrame({'i': list(range(1, k+1))})
interval_df["[x_i, x_i+1)"] = intervals
interval_df['Середина i интервала'] = mid_of_intervals
interval_df['n_i'] = abs_freq
interval_df['p*_i'] = rel_freq
interval_df['Накопленная n_i'] = accum_abs_freq
interval_df['Накопленная p*_i'] = accum_rel_freq

print('\nТаблица 1:')
# Отображение Таблицы 1
df_styler = interval_df.style.set_properties(**{'text-align': 'center'})
df_styler.hide(axis='index')
```

### *__Задание 5__*
*Для абсолютных частот построить и отобразить графически полигон, гистограмму и эмпирическую функцию. Сделать выводы.*

- Построим *__полигон абсолютных частот__*.
> *__Полигон абсолютных/относительных частот__* - ломаная, соединяющая точки, соответствующие по оси Ox - серединам интервалов интервального ряда, по оси Oy - абсолютным/относительным частотам этих интервалов. Площадь таких прямоугольников равна абсолютной/относительной частоте. <br>

- Построим *__гистограмму абсолютных частот__*.
> *__Гистограмма абсолютных/относительных частот__* - фигура, состоящая из прямоугольников, основаниями которых служат интервалы интервального ряда, а высоты равны отношению абсолютных/относительных частот интервала к его ширине.

- Построим *__эмпирическую функцию распределения__*:

> *__Эмпирической функцией распределения__* (функцией распределения выборки) называют функцию $F^*(x)$, определяющую для каждого значения $x$ относительную частоту события $X < x$.

&emsp;График $F^*(x)$ представляет собой лестничный график, длина каждой ступеньки которого равна длине соответствующего интервала, а высота - отношению накопленной частоты до середины этого интервала к объему выборки, т.е.
$$F^*( \widetilde{x}_i) = \frac{m_i^{нак}}{N}; \\  m_i^{нак} = \sum_{j=1}^{i-1}m_j; \\ i=1,2,...,k+1$$

```python
# Размеры графика
plt.figure(figsize=[12, 4])

# Построение полигона абсолютных частот
# Отложим на осях необходимые для полигона частот точки
x = mid_of_intervals # x - Середины интервалов
y = abs_freq # y - Абсолютные частоты
plt.subplot (2, 2, 1)
plt.plot(x, y, marker='o')
plt.title ("Полигон абсолютных частот")
plt.xticks(x)
plt.yticks([i for i in range(1, 25, 2 )])
plt.xlabel(r'Середина $\tilde{x_i}$')
plt.ylabel(r'Абс.частота $n_i$')
plt.grid()

# Построение гистограммы абсолютных частот
# Отложим на осях необходимые для гистограммы частот точки
x = mid_of_intervals # x - Середины интервалов
# Высоты прямоугольников гистограммы
y = []
for i in range(k):
  y.append(abs_freq[i] / h)  # Отношение абсолютной частоты к ширине интервала h

plt.subplot (2, 2, 3)
plt.xticks(mid_of_intervals)
plt.title('Гистограмма абсолютных частот')
plt.xlabel(r'Середина $\tilde{x_i}$')
plt.ylabel(r'Абс.частота $n_i$ / Ширина h')
plt.grid()
plt.bar(mid_of_intervals, y, h, color='green')

# Построение эмпирической функции распределения
plt.subplot (1, 2, 2)

x = []
x.append(mid_of_intervals[0] - h)
x.append(mid_of_intervals[0])

for i in range(k):
  x.append(mid_of_intervals[i])
  x.append(mid_of_intervals[i]+ h)

x.append(mid_of_intervals[-1] + h)

y = [0, 0]
for i in range(k):
  y.append(accum_abs_freq[i]/variant_count)
  y.append(accum_abs_freq[i]/variant_count)

y.append(1)

plt.xticks(x[1:-2])
plt.yticks(y)
plt.title('Эмпирическая функция распределения')
plt.xlabel(r'Середина $\tilde{x_i}$')
plt.ylabel(r'Накопленная частота $n^\Sigma_{i}/N$')

plt.plot(x, y)
plt.subplots_adjust(hspace=1)

plt.grid()
plt.show()
```
