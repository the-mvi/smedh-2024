## Основные теоретические положения

**Метод наименьших квадратов (МНК)** - метод, основанный на поиске минимума суммы квадратов отклонений значений некоторых функций от заданного множества значений. МНК является одним из основных методов регрессионного анализа и применяется для оценки параметров регрессионных моделей на основе выборочных данных

$$M(X/y)=q_1(y)$$
$$M(Y/x)=q_2(x)$$
****
Пусть имеется двумерная случайная величина $\{X,Y\}$, где $X$ и $Y$ зависимые случайные величины.

И $Y$ как линейную функцию случайной величины $X$:
$$Y ≃ g(x) = ax + b$$
Значения коэффициентов $a$ и $b$ определяются с помощью МНК из условия минимума функции
$$F(a,b) = m[(Y-aX-b)^2]$$
В этом случае функцию $g(x)$ называют **линейной функцией среднеквадратической регрессии $Y$ на $X$**
$$g(x) = m(Y/x) = m(Y)+r_{xy}\frac{σ_y}{σ_x}[x-m(X)]$$

Коэффициент $ρ_{yx}=r_{xy}\frac{σ_y}{σ_x}$ называют **коэффициентом регрессии $Y$ на $X$**

Минимальное значение функции $F(a,b)$ равно $σ_y^2(1-r_{xy}^2)$ и называется **остаточной дисперсией случаной величины $Y$ относительно случайной величины $X$**

Линейная функция среднеквадратической регрессии $X$ на $Y$ строится аналогично:
$$g(y) = m(X/y) = m(X)+r_{xy}\frac{σ_x}{σ_y}[y-m(Y)]$$
$ρ_{xy}=r_{xy}\frac{σ_x}{σ_y}$ - коэффициент регрессии $X$ на $Y$, остаточная дисперсия равна $σ_x^2(1-r_{xy}^2)$

****
В случае, когда известны только выборочные данные - двумерная выборка значений случайных величин X и Y, возможно построение только выборочных прямых среднеквадратической регрессии. Уравнения выборочных прямых среднеквадратической регрессии получаются на основе уравнений линейных среднеквадратичных регрессий, записанных ранее, в которых все параметры распределений заменяются их статистическими оценками:
$$\bar y_x = \bar y_в + \bar r_{xy}\frac{S_y}{S_x}(x-\bar x_в)$$
$$\bar x_y = \bar x_в + \bar r_{xy}\frac{S_x}{S_y}(y-\bar y_в)$$
****
Оценку **общей дисперсии** $X$ можно представить как сумму внутригрупповой и межгрупповой дисперсий:
$$D_{общ}=D_{межгр}+D_{внгр}$$
где $D_{внгр}$ - **внутригрупповая дисперсия**, определяемая как взвешенной по объемам групп средняя арифметическая групповых дисперсий

$D_{межгр}$ - **межгрупповая дисперсия**, определяемая как дисперсия условных (групповых) средних $\bar x_{y_i}$ относительно выборочной средней $\bar x_в$

Если $D_{межгр} = 0$, то значит, что все условные (групповые) средние равны и не зависят от выборочных значений $y$. В этом случае выборочные данные не дают оснований предположить, что между исследуемыми случайными величинами имеет место корреляционная зависимость.

Если $D_{внгр} = 0$, то значит, что в каждой из групп имеется только одно значение $X$. В этом случае выборочные данные позволяют предположить наличие функциональной зависимости между исследуемыми случайными величинами

****
**Выборочное корреляционное отношение $X$ к $Y$** определяется выражением:
$$\bar \eta_{xy} = \frac{\bar \sigma_{\bar x_y}}{\bar \sigma_x}$$
где $\bar \sigma _{\bar x_y} = \sqrt{D_{межгр}}$

$\bar \sigma_x = \sqrt{D_{общ}}$ - выборочные значения СКВО $\bar x_y$ и $X$ соответсвенно

Аналогично определяется выборочное корреляционное отношение $Y$ к $X$:
$$\bar \eta_{yx} = \frac{\bar \sigma_{\bar y_x}}{\bar \sigma_y}$$
как для $\bar \eta_{xy}$ так и для $\bar \eta_{yx}$ имеет место равенство:
$$D_{внгр} = D_{общ}(1-\frac{D_{межгр}}{D_{общ}}) = D_{общ}(1-\eta^2)$$
****
Свойства выборочного корреляционного отношения (имеют место как для $\bar \eta_{xy}$ так и для $\bar \eta_{yx}$):


1.   $0 ≤ \bar \eta ≤ 1$
2.   $\bar \eta = 0$ - выборочные данные согласованы с предположением, что случайные величины $X$ и $Y$ не связаны корреляционной зависимостью
3. $\bar \eta = 1$ - выборочные данные согласованы с предположением, что случайные величины $X$ и $Y$ связаны функциональной зависимостью
4. $\bar \eta \geq |\bar r_{xy}|$
5. $\bar \eta = |\bar r_{xy}|$ - выборочные данные согласованы с предположением, что случайные величины  $X$ и $Y$ связаны линейной корреляционной зависимостью
6. Корреляционное отношение является количественной мерой тесноты корреляционной зависимости между случайными величинами  $X$ и $Y$. Вместе с тем, кроме п.5, оно не позволяет определить характер корреляционной зависимости
****
Пусть выборочное уравнение регрессии $Y$ на $X$ имеет слудющий вид:
$$\bar y_x = ax^2 +bx + c$$
Значения коэффициентов $a,b$ и $c$ можно определить с помощью МНК, что приводит к необходимости решения системый линейных уравнений третьего порядка:

То есть получается следующая система уравнений:
\begin{equation}
\left\{
\begin{array}{ll}
0.000003598589863420782a+0.005531920385903251b = 0.34969786630291944\\
(0.005531920385903251a + 108b = 584.2797923209687\end{array}\right.
\end{equation}

```python
##############################
#Решение системы с помощью numpy


left_side = np.array([[0.000003598589863420782, 0.005531920385903251],
  [0.005531920385903251, 108]])

right_side = np.array([0.34969786630291944, 584.2797923209687])

a,b = np.linalg.inv(left_side).dot(right_side)

print("a = " + str(a))
print("b = " + str(b))
```
