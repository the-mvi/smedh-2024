## Основные теоретические положения

**Генеральная совокупность** - совокупность всех объектов, из которых производится выборка
****
**Выборка** - совокупность случайно отобранных элементов генеральной совокупности

**Главный принцип формирования репрезентативной выборки** - случайность ее формирования

Выборки бывают **повторные** и **бесповторные** (если случайно отбираемые объекты не возвращаются в генеральную совокупность, то это бесповторная выборка. Если же выбранный объект возвращается обратно (перед выбором следующего), то это повторная выборка)

По способу формирования выборки разделяются на две группы:

1) **Без разбиения генеральной совокупности на части**:

*   Простой случайный бесповторный отбор (простой случайный отбор - случайным образом отбираются элементы генеральной совокупности)
*   Простой случайный повторный отбор

2) **С разбиением генеральной совокупности на части:**

*   Типический (генеральная совокупность разделяется на несколько подгрупп в соответствии с каким-либо признаком, а затем производится случайный отбор в каждой подгруппе пропорциональной доли данной подгруппы от всей генеральной совокупности)
*   Механический (генеральная совокупность разделятся на несколько равных (по возможности) интервалов, затем из первого интервала случайно отбирается один элемент, а потом с шагом, равным длине интервала, отбираются другие элементы из последующих интервалов)
*   Серийный (генеральная совокупность разделяется на несколько групп (серий) таким образом, чтобы в каждой группе было поровну (по возможности) элементов; затем случайным образом выбирается одна серия, а затем при помощи простого, типического или механического отбора отбирается определенное числое элементов)
****
**Статистический ряд** - последовательность элементов выборки, расположенных в порядке их получения (наблюдения)

****
**Ранжированный ряд** - последовательность элементов выборки, расположенных в порядке возрастания их значений. Номер элемента ранжированного ряда в последовательности называется **рангом**
****
**Вариационный ряд** - получается из ранжированного ряда в результате объединения одинаковых элементов. Элементы вариационного ряда называются **вариантами**.
****
**Построение интервального ряда:**

Если количество вариант $K$ достаточно велико, то необходимо разделить вариационный ряд на интервалы. В этом случае ряд называется **интервальным**. Количество интервалов $k$ зависит от объема выборки $N$ и может быть вычислено по формуле Стерджесса:
$$k = 1+3.31*lgN$$
Полученное значение округляется до целого. Рекомендуется выбирать нечетное количество интервалов. Затем вычисляется ширина интервалов $h$ по формуле:
$$h = \frac{x_{max}-x_{min}}{k}$$

Границы интервалов определяются по формуле:
$$[x_{min}+(i-1)*h; x_{min}+i*h), i = 1,2,...,k-1,$$ где $i$ - номер интервала

Последний интервал определяется по формуле:
$$[x_{min} + (k-1)*h; x_{min}+k*h]$$
****
**Абсолютная частота в интервальном ряде** - количество вариант, попавших в определенный интервал интервального ряда
****
**Относительная частота в интервальном ряде** - отношение абсолютной частоты к общему количеству вариант
****
**Полигон абсолютных частот** - ломаная, соединяющая точки, соответствующие срединным значениям интервалов интервального ряда и абсолютным частотам этих интервалов (**полигон относитльных частот** является такой же ломаной, но вместо абсолютных частот отмечаются относительные частоты).
****
**Гистограмма абсолютных частот** - фигура, состоящая из прямоугольников, основаниями которых служат интервалы интервального ряда, а высоты равны отношению абсолютных частот интервала к его ширине (**гистограмма относительных частот** является такой же фигурой, но вместо абсолютных частот отмечаются относительные частоты).
****
**Эмпирическая функция распределения** называют функцию $F^*(x)$, определяющую для каждого значения $x$ относительную частоту событий $X < x$.

График $F^*(x)$ представляет собой лестничный график, длина каждой ступеньки которого равна длине соответствующего интервала, а высота - отношению накопленной частоты до середины этого интервала к объему выборки, т.е.
$$F^*( \widetilde{x}_i) = \frac{m_i^{нак}}{N}; m_i^{нак} = \sum_{j=1}^{i-1}m_j; i=1,2,...,k+1$$

Вариационные ряды преобразуются в интервальные ряды в виде таблиц (преобразования описаны в теоритических положениях, а также к комментариях к коду)

```python
def get_interval_series(variational_series):
  # возвращает список интервалов, принимает вариационный ряд

  # формула Стерджесса
  k = round(1 + 3.31 * math.log10(N))
  # количество интервалов (т.к.
  #рекомендуется выбирать нечетное, то в случае четного добавляется 1)
  if k % 2 == 0:
    k += 1

  max_elem = max(variational_series)
  min_elem = min(variational_series)
  h = (max_elem - min_elem) / k # ширина интервалов

  interval_serires = []

  for i in range(1, k): # границы интервалов
    left_border = min_elem + (i - 1) * h
    right_border = min_elem + i * h
    interval_serires.append([left_border, right_border])

  left_border = min_elem + (k - 1) * h # последний интервал
  right_border = min_elem + k * h
  interval_serires.append([left_border, right_border])

  return interval_serires

def get_frequencies(interval_serires, variational_series):
  # возвращает списки абсолютных и относительных частот для интервального ряда,
  # принимет интервальный и вариационный ряды

  size = len(interval_serires)
  absolute_frequencies = [0] * size

  for variant in variational_series:
    for i in range(size):
      left_border = interval_serires[i][0]
      right_border = interval_serires[i][1]
      if (i+1) != size:
        if variant >= left_border and variant < right_border:
          absolute_frequencies[i] += 1
      else:
        if variant >= left_border and variant <= right_border:
          absolute_frequencies[i] += 1

    relative_frequencies = [i / len(variational_series) for i in
                            absolute_frequencies]

  return absolute_frequencies, relative_frequencies

def display_table(interval_serires, frequencies):
  # визализирует интервальный ряд в виде таблицы,
  # принимает интервальный ряд, список списков абсолютных и относительных частот

  size = len(interval_serires)
  myTable = PrettyTable(["Границы интервалов", "Середины интервалов",
        "Частота попадания в интервал", "Относительная частота"])

  for i in range(len(interval_serires) - 1):
    left_border = interval_serires[i][0]
    right_border = interval_serires[i][1]
    interval_str = "[" + str(left_border) + ", " + str(right_border) + ")"
    middle_interval_str = str((left_border + right_border) / 2)
    absolute_frequency_str = str(frequencies[0][i])
    relative_frequencies_str = str(frequencies[1][i])

    myTable.add_row([interval_str, middle_interval_str, absolute_frequency_str,
                     relative_frequencies_str])

  left_border = interval_serires[size-1][0]
  right_border = interval_serires[size-1][1]
  interval_str = "[" + str(left_border) + ", " + str(right_border) + "]"
  middle_interval_str = str((left_border + right_border) / 2)
  absolute_frequency_str = str(frequencies[0][size-1])
  relative_frequencies_str = str(frequencies[1][size-1])

  myTable.add_row([interval_str, middle_interval_str, absolute_frequency_str,
                     relative_frequencies_str])

  return myTable

first_interval_serires = get_interval_series(first_sampling)
second_interval_series = get_interval_series(second_sampling)

first_frequencies = get_frequencies(first_interval_serires, first_sampling)
second_frequencies = get_frequencies(second_interval_series, second_sampling)

print("Для первого столбца")
print(display_table(first_interval_serires, first_frequencies))
print("Для второго столбца")
print(display_table(second_interval_series, second_frequencies))
```

Для обоих столбцов строятся полигоны для обеих частот, на оси абсцисс откладываются середины интервалов, на оси ординат - частоты

```python
def _get_middle_intervals(interval_serires): # вспомогательная функция
  # возвращает списков середин интервалов, принимает интервальный ряд

  middle_intervals = []
  for interval in interval_serires:
      left_border = interval[0]
      right_border = interval[1]
      middle_intervals.append((left_border + right_border) / 2)

  return middle_intervals


def display_polygons(frequencies, interval_serires):
  # визулизирует полигоны для абсолютных и относительных частот
  # принимает список списков абсолютных и относительных частот и
  # интервальный ряд

  x = _get_middle_intervals(interval_serires)
  y = frequencies[0]

  fig, ax = plt.subplots()
  poly = Polygon(list(zip(x, y)), facecolor='none', edgecolor='blue',
               closed=False)
  ax.add_patch(poly)
  plt.plot(x, y, marker='o', markersize=5, color='blue')
  ax.set_xlim([interval_serires[0][0],
             interval_serires[len(interval_serires)-1][1]])
  ax.set_ylim([0, 1.5*max(frequencies[0])])
  plt.xticks(x)
  plt.title('Полигон для абсолютных частот')
  plt.xlabel('Середина интервала')
  plt.ylabel('Частота')
  plt.grid()
  plt.show()

  plt.clf()
  y = frequencies[1]
  poly = Polygon(list(zip(x, y)), facecolor='none', edgecolor='blue',
               closed=False)
  ax.add_patch(poly)
  plt.plot(x, y, marker='o', markersize=5, color='blue')
  ax.set_xlim([interval_serires[0][0],
             interval_serires[len(interval_serires)-1][1]])
  ax.set_ylim([0, 1.5*max(frequencies[0])])
  plt.xticks(x)
  plt.title('Полигон для относительных частот')
  plt.xlabel('Середина интервала')
  plt.ylabel('Частота')
  plt.grid()
  plt.show()

print("Для первого столбца:")
display_polygons(first_frequencies, first_interval_serires)
```

```python
print("Для второго столбца:")
display_polygons(second_frequencies, second_interval_series)
```

Для обоих столбцов строятся гистограммы для обеих частот, на оси абсцисс отклываются середины интервалов, на оси ординат - отношение частоты и ширины интервалов. Площадь выводимых фигур равна соответствующей частоте

```python
def _get_hist_data(frequencies, interval_serires): # вспомогательная функция
  # возвращает середины интервалов, ширину и отношения частоты и ширины интервалов

  height = []
  left_border = interval_serires[0][0]
  right_border = interval_serires[0][1]
  width = right_border - left_border
  for i in range(len(interval_serires)):
    height.append(frequencies[i]/width)
  middle_intervals = _get_middle_intervals(interval_serires)
  return middle_intervals, width, height

def display_hist(frequencies, interval_serires, width_correct):
  # визуализирует гистограммы для обеих частот
  # принимает список списков частот, интервальный ряд и корректировку для фигур
  # (чтобы они не сливались - т.е. между ними были границы)

  middle_intervals, width, height = _get_hist_data(frequencies[0],
                                                   interval_serires)
  ax = plt.axes()
  ax.set_axisbelow(True)
  plt.xticks(middle_intervals)
  plt.title('Гистограмма для абсолютных частот')
  plt.xlabel('Середина интервала')
  plt.ylabel('Частота/Ширина')
  plt.grid()
  plt.bar(middle_intervals, height, width-width_correct, color='blue')
  plt.show()

  plt.clf()
  middle_intervals, width, height = _get_hist_data(frequencies[1],
                                                   interval_serires)
  ax = plt.axes()
  ax.set_axisbelow(True)
  plt.xticks(middle_intervals)
  plt.title('Гистограмма для относительных частот')
  plt.xlabel('Середина интервала')
  plt.ylabel('Частота/Ширина')
  plt.grid()
  plt.bar(middle_intervals, height, width-width_correct, color='blue')
  plt.show()

print("Для первого столбца:")
display_hist(first_frequencies, first_interval_serires, 4)
```

```python
print("Для второго столбца:")
display_hist(second_frequencies, second_interval_series, 0.5)
```

Для обоих столбцов строится эмпирическая функция распределения. В силу формулы для вычисления эмпирической фунции, эмпирическая функция для обеих частот совпадает.

```python
def display_ECDF(interval_serires, sampling, frequencies):
  m = frequencies[0]
  accumulated_sum = []
  for i in range(len(m)):
    accumulated_sum_i = sum(m[:i])
    accumulated_sum.append(accumulated_sum_i)

  empirical_F_values = [x / len(sampling) for x in accumulated_sum]

  x = []
  middle_intervals = _get_middle_intervals(interval_serires)
  width = middle_intervals[1] - middle_intervals[0]
  x.append(middle_intervals[0] - width)
  x_len = len(middle_intervals)
  for i in range(x_len):
    x.append(middle_intervals[i])
    x.append(middle_intervals[i])
  x.append(middle_intervals[x_len-1]+width)

  y = []
  y_len = len(empirical_F_values)
  y.append(empirical_F_values[0])
  for i in range(y_len-1):
    y.append(empirical_F_values[i])
    y.append(empirical_F_values[i+1])

  y.append(empirical_F_values[y_len-1])
  y.append(1)
  y.append(1)

  fig, ax = plt.subplots()
  poly = Polygon(list(zip(x, y)), facecolor='none', edgecolor='blue',
               closed=False)
  ax.add_patch(poly)
  plt.plot(x, y, color='blue')
  plt.xticks(middle_intervals)
  plt.yticks(y)
  plt.title('Эмпирическая функция распределения')
  plt.xlabel('Середина интервала')
  plt.ylabel('m_i нак / n')
  plt.grid()
  plt.show()

print("Для первого интервального ряда")
display_ECDF(first_interval_serires, first_sampling, first_frequencies)
print("Для второго интервального ряда")
display_ECDF(second_interval_series, second_sampling,second_frequencies)
```

**Выводы по пункту:**

1) Как можно заметить, графики для абсолютных и относительных частот либо прямо пропорциональны (в случае полигонов и гистограмм) или совпадают (в случае эмпирических функций). Это следует из того, что относительная частота вычисляется на основе абсолютной путем деления на общее количество вариант

2) Полигоны и гистограммы подтверждают выводы сделанные в предыдущем пункте насчет того, что больше всего значений для первого интервального ряда концентруется ближе к серединным интервалам (за исключением небольшого скачка примерно в середине графика), а графики для второго интервального ряда демонстрируют возрастание частоты значений между начальным и серединным интервалом, и непрерывное уменьшение частоты до конечного интервала

3) Графики эмпирических функций распределения для обоих рядов демонстрируют, что накопленная частота в начале сильно возрастает в области центральных значений середин интервалов, но ближе к концу возрастание постепенно снижается (в случае первого ряда заметно, что возрастание снижается более резко, чем в случае второго ряда)
